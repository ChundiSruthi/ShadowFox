```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from transformers import pipeline
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
```

```python
result = classifier("I loved this movie. It was amazing!")

print(result)
```

```python
result = classifier("This was the worst product I have ever used.")

print(result)
```

```python
reviews = [
    "I absolutely loved the performance.",
    "The movie was boring and too long.",
    "The product quality is excellent.",
    "I am very disappointed with the service.",
    "The experience was okay, not great."
]

results = classifier(reviews)

for review, result in zip(reviews, results):
    print("Review:", review)
    print("Prediction:", result)
    print()
```

```python
labels = [result['label'] for result in results]

positive_count = labels.count('POSITIVE')
negative_count = labels.count('NEGATIVE')

plt.bar(['Positive', 'Negative'], [positive_count, negative_count])

plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Sentiment Analysis Results")

plt.show()
```
